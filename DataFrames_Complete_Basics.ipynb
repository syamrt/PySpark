```
from pyspark.sql.types import IntegerType
def total_salary(x,y):
    return x+y;

total_salaryUDF = udf(lambda x,y: total_salary(x,y), IntegerType())

office_data.withColumn('total_salary', total_salaryUDF(office_data.salary, office_data.bonus)).show()
```

```
from pyspark.sql.types import FloatType
def increment(salary, bonus, state):
    if state=='NY':
        return (0.05*bonus) + (0.1*salary)
    elif state=='CA':
        return (0.03*bonus) + (0.12*salary)

incrementUDF = udf(lambda x,y,z: increment(x,y,z), FloatType())

office_data.withColumn("increment", incrementUDF(office_data.salary, office_data.bonus, office_data.state)).show()
```
